```python
def reduce_memory_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Starting memory usage: {:5.2f} MB'.format(start_mem))

    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min >= np.iinfo(np.int32).min and c_max <= np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min >= np.iinfo(np.int64).min and c_max <= np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min >= np.finfo(np.float16).min and c_max <= np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min >= np.finfo(np.float32).min and c_max <= np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Reduced memory usage: {:5.2f} MB ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```python
#df_cicle[::100]['signal'].rolling(window=50, center=True).apply(lambda x: stats.mode(x)[0])
def get_top3_1(x):
    unique, counts = np.unique(x, return_counts=True)
    top3 = unique[np.argsort(counts)][-3:]
    top3.sort()
    return top3[0]

def get_top3_2(x):
    unique, counts = np.unique(x, return_counts=True)
    top3 = unique[np.argsort(counts)][-3:]
    top3.sort()
    return top3[1]

def get_top3_3(x):
    unique, counts = np.unique(x, return_counts=True)
    top3 = unique[np.argsort(counts)][-3:]
    top3.sort()
    return top3[2]

mc1 = df_cycle[::100]['signal'].rolling(window=50, center=True).apply(get_top3_1)
mc2 = df_cycle[::100]['signal'].rolling(window=50, center=True).apply(get_top3_2)
mc3 = df_cycle[::100]['signal'].rolling(window=50, center=True).apply(get_top3_3)
df_mc_top3 = pd.DataFrame([mc1,mc2,mc3]).transpose()
df_mc_top3.columns = ['mc1', 'mc2', 'mc3']
```

```python
from bisect import bisect_left

def takeClosest(myList, myNumber):
    """
    Assumes myList is sorted. Returns closest value to myNumber.
    If two numbers are equally close, return the smallest number.
    """
    pos = bisect_left(myList, myNumber)
    if pos == 0:
        return myList[0]
    if pos == len(myList):
        return myList[-1]
    before = myList[pos - 1]
    after = myList[pos]
    if after - myNumber < myNumber - before:
        return after
    else:
        return before

new_signal = df_cycle_red['signal'].apply(lambda x: takeClosest(points,x))
```

```python
print("Using RLE compression the signal could be reduced to approx. {:.2f}% of the original size".format(len(np.where(np.diff(new_signal) != 0)[0])/len(new_signal)*100))
```

```python
def rle_encode(values):
    difs = np.diff(values)
    jumps = np.where(difs != 0)[0]
    if len(jumps)==0:
        return np.array([values[0], [0]]).transpose()
    pos = np.append([0], jumps+1)
    return np.array([pos, values[pos]]).transpose()
#    lengths = [-1] 
#    lengths.extend(jumps)
#    lengths.extend([len(values)-1])
#    lengths = np.diff(lengths)
#    return {'value': values[pos], 'pos': pos, 'rep': lengths}    

#rle_encode(np.array([1,1,1,2,2,1,3,3,3]))
rle = rle_encode(new_signal.values)

plt.figure(figsize=(12, 4))
plt.plot(rle[:,0], rle[:,1])
plt.show()
```

```python
import operator

def hausdorff_peaks(x1, x2, points):
    dist = 0
    for p in points:
        x1p = x1[x1[:,1]==p]
        x2p = x2[x2[:,1]==p]
        if (len(x1p)!=0 and len(x2p)!=0):
            dist += abs(p)*max(directed_hausdorff(x2p, x1p)[0], directed_hausdorff(x1p, x2p)[0]) 
        elif (len(x1p)!=0 or len(x2p)!=0):
            dist += abs(p)**2.5
    return dist
    

def getNeighborsTimes_HaussPeaks(trainingSet, testInstance, k, points):
    distances = []
    length = len(testInstance)-1
    for x in range(len(trainingSet)):
        distances.append((x, hausdorff_peaks(trainingSet[x], testInstance, points)))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors

#getNeighborsTimes_HaussPeaks(train_X, train_X[10], 3, points)
```

Although it might represent well the signal, the last distance used will perform very poorly pairing segments with high peaks but which are not exactly of the same heigh.

```python
from bisect import bisect_left

def takeClosest(myList, myNumber):
    """
    Assumes myList is sorted. Returns closest value to myNumber.
    If two numbers are equally close, return the smallest number.
    """
    pos = bisect_left(myList, myNumber)
    if pos == 0:
        return myList[0]
    if pos == len(myList):
        return myList[-1]
    before = myList[pos - 1]
    after = myList[pos]
    if after - myNumber < myNumber - before:
        return after
    else:
        return before

new_signal = df_train['signal'].apply(lambda x: takeClosest(points,x))
```

```python
def rle_encode(values):
    difs = np.diff(values)
    jumps = np.where(difs != 0)[0]
    if len(jumps)==0:
        return np.array([values[0], [0]]).transpose()
    pos = np.append([0], jumps+1)
    return np.array([pos, values[pos]]).transpose() 

rows = 150000
segments = int(np.floor(df_train.shape[0]/rows))

train_X = pd.DataFrame(columns=['points', 'values', 'seg_id'], dtype=np.int16)
train_y = pd.DataFrame(index=range(segments), columns=['seg_id', 'time'])
for seg_id in range(segments):
    seg = df_train.iloc[seg_id*rows:seg_id*rows+rows]
    rle = rle_encode(seg['new_signal'].values)
    rle = np.hstack((rle, seg_id*np.ones((rle.shape[0],1)))) #Add the seg_id
    df_rle = pd.DataFrame(rle, columns=train_X.columns)
    train_X = train_X.append(df_rle)
    train_y.loc[seg_id] = [seg_id, seg['time'].values[-1]]
```
